```python
import numpy as np
import pandas as pd
from matplotlib import pyplot

#Preapre transition matrix and initial state
tm = np.matrix(
    [
        [0.82, 0.18, 0.00],
        [0.00, 0.47, 0.53],
        [0.10, 0.00, 0.90],
    ]
)
s1 = np.array([1,0,0])
```

```python
#Build Markov Chain
iterations = 35
plot_data = []
for step in range(iterations):
    result = s1 * tm ** step
    plot_data.append(np.array(result).flatten())
```

```python
#Convert to np.array
plot_data = np.array(plot_data)
```

```python
#Plotting and saving
pyplot.figure(1)
pyplot.xlabel("Steps")
pyplot.ylabel("Probability")
pyplot.title("Evolution of the generic epidemics")
lines = []
for i, shape in zip(range(3), ["-x", "-h", "-H"]):
    y = pyplot.plot(plot_data[:, i], shape)
    lines.append(y)
pyplot.legend(["Susceptible","Infected","Removed"])
#pyplot.savefig('Evolution.png')
pyplot.show()
```
